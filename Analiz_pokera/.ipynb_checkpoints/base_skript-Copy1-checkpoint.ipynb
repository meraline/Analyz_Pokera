```python
%%time

data_pd = {'HandId':[], 'Tournament':[], 'Level':[], 'Date':[], 'Hour':[], 
           'TimeZone':[], 'Table':[], 'HandHiro':[], 'Board':[], 'Players':[], 'pref_call':[], 
           'pref_raises':[], 'pref_allin':[], 'flop_call':[], 'flop_check':[], 'flop_bet':[], 'flop_raises':[], 
           'turn_call':[],'turn_check':[], 'turn_bet':[], 'turn_raises':[], 'river_call':[], 'river_check':[], 
           'river_bet':[], 'river_raises':[], 'showed_won':[]}

with open('Poker_/SNG_452883.txt', 'r', encoding="utf8") as f:data = f.read() # Poker_\SNG dimetruk 1-54133.txt Poker_\SNG dimetruk 1-14437  dim1000.txt
    
#result = re.findall('(\w+) Hand #(\d+): (.*) \(.(\d+\.\d+)\/.(\d+\.\d+).*(\d\d\d\d\/\d\d\/\d\d).*(\d+:\d+:\d+) (\w+)', data)
result = re.findall('(\w+) Hand #(\d*): Tournament #(\d*),(.*) Level (.*) - (\d{4}/\d+/\d+) (\d+\D\d+\D\d+) (\w*)', data)
#rakes = re.findall('Rake \$(.*)', data)
Tables = re.findall('Table \'.*\' (.*) Seat .*\n', data)
#bains = re.findall('(\D\d+\.\d+\+\D\d+\.\d+\+\D\d+\.\d+)', data)
HandHiros = re.findall("Dealt to (.*) \D(.. ..)\D", data)
Boards = re.findall("Board \[(.*)\]", data)


s = re.findall('Seat (\d).*chips', data)
s = [int(x) for x in s]
seats = []
length = 1
for i in range(len(s)-1):
    if s[i+1] < s[i]:
        seats.append(length)
        length = 1
    else:
        length += 1
seats.append(length)

pref_calls = []
pref_raisess= []
pref_allins = []

flop_calls = []
flop_checks = []
flop_bets = []
flop_raisess= []

turn_calls = []
turn_checks = []
turn_bets = []
turn_raisess= []

river_calls = []
river_checks = []
river_bets = []
river_raisess= []

showed_wons = []

length_calls = 0
length_raisess = 0
length_allins = 0

length_f_calls = 0
length_f_checks = 0
length_f_bets = 0
length_f_raisess = 0

length_t_calls = 0
length_t_checks = 0
length_t_bets = 0
length_t_raisess = 0

length_r_calls = 0
length_r_checks = 0
length_r_bets = 0
length_r_raisess = 0

for i in range(len(data.split("\n\n\n"))): # -1 уход от ошибки, выход из цикда без последнего элемента
    try:
        distribut = data.split("\n\n\n")[i]
        hhh = PokerStarsHandHistory(distribut)
        hhh.parse()
        
        pfa = str(hhh.preflop_actions)
        fa = str(hhh.flop.actions)
        ta = str(hhh.turn_actions)
        ra = str(hhh.river_actions)

        pr_call = re.findall("calls", pfa)
        pr_rais = re.findall("raises", pfa)
        pr_allin = re.findall("all-in", pfa)

        f_check = re.findall("check", fa)
        f_call = re.findall("call", fa)
        f_raise = re.findall("raise", fa)
        f_bet = re.findall("bet", fa)
        #print(f_check,f_call)
        t_check = re.findall("check", ta)
        t_call = re.findall("call", ta)
        t_raise = re.findall("raise", ta)
        t_bet = re.findall("bet", ta)
        
        r_check = re.findall("check", ra)
        r_call = re.findall("call", ra)
        r_raise = re.findall("raise", ra)
        r_bet = re.findall("bet", ra)
# префлоп действия
        if pr_call == []:
            length_calls = 0
        else:
            for c in range(len(pr_call)):
                if c==0:
                    length_calls = 1
                else:
                    length_calls += 1  
        pref_calls.append(length_calls)

        if pr_rais == []:
            length_raisess = 0
        else:
            for r in range(len(pr_rais)):
                if r==0:
                    length_raisess = 1
                else:
                    length_raisess += 1  
        pref_raisess.append(length_raisess)

        if pr_allin == []:
            length_allins = 0
        else:
            for a in range(len(pr_allin)):
                #print(a)
                if a==0:
                    length_allins = 1
                else:
                    length_allins += 1  
        pref_allins.append(length_allins)
# флоп действия
        if f_check == []:
            length_f_checks = 0
        else:
            for ch in range(len(f_check)):
                if ch==0:
                    length_f_checks = 1
                else:
                    length_f_checks += 1  
        flop_checks.append(length_f_checks)
        
        if f_call == []:
            length_f_calls = 0
        else:
            for k in range(len(f_call)):
                if k==0:
                    length_f_calls = 1
                else:
                    length_f_calls += 1  
        flop_calls.append(length_f_calls)
        
        if f_bet == []:
            length_f_bets = 0
        else:
            for b in range(len(f_bet)):
                if b==0:
                    length_f_bets = 1
                else:
                    length_f_bets += 1  
        flop_bets.append(length_f_bets)
        
        if f_raise == []:
            length_f_raisess = 0
        else:
            for fr in range(len(f_raise)):
                if fr==0:
                    length_f_raisess = 1
                else:
                    length_f_raisess += 1  
        flop_raisess.append(length_f_raisess)
# тёрн действия        
        if t_check == []:
            length_t_checks = 0
        else:
            for tch in range(len(t_check)):
                if tch==0:
                    length_t_checks = 1
                else:
                    length_t_checks += 1  
        turn_checks.append(length_t_checks)
        
        if t_call == []:
            length_t_calls = 0
        else:
            for tk in range(len(t_call)):
                if tk==0:
                    length_t_calls = 1
                else:
                    length_t_calls += 1  
        turn_calls.append(length_t_calls)
        
        if t_bet == []:
            length_t_bets = 0
        else:
            for tb in range(len(t_bet)):
                if tb==0:
                    length_t_bets = 1
                else:
                    length_t_bets += 1  
        turn_bets.append(length_t_bets)
        
        if t_raise == []:
            length_t_raisess = 0
        else:
            for tr in range(len(t_raise)):
                if tr==0:
                    length_t_raisess = 1
                else:
                    length_t_raisess += 1  
        turn_raisess.append(length_t_raisess)
# ривер действия
        if r_check == []:
            length_r_checks = 0
        else:
            for rch in range(len(r_check)):
                if rch==0:
                    length_r_checks = 1
                else:
                    length_r_checks += 1  
        river_checks.append(length_r_checks)
        
        if r_call == []:
            length_r_calls = 0
        else:
            for rk in range(len(r_call)):
                if rk==0:
                    length_r_calls = 1
                else:
                    length_r_calls += 1  
        river_calls.append(length_r_calls)
        
        if r_bet == []:
            length_r_bets = 0
        else:
            for rb in range(len(r_bet)):
                if rb==0:
                    length_r_bets = 1
                else:
                    length_r_bets += 1  
        river_bets.append(length_r_bets)
        
        if r_raise == []:
            length_r_raisess = 0
        else:
            for rr in range(len(r_raise)):
                if rr==0:
                    length_r_raisess = 1
                else:
                    length_r_raisess += 1  
        river_raisess.append(length_r_raisess)
        
        show_wo = re.findall("Seat (\d+): (.+?) showed \[(.*)\] and won", distribut) # карты победителя
        #print(show_wo[0][2])
        if show_wo != []:
            showed_wons.append(show_wo[0][2])
        else:
            showed_wons.append(np.NaN)
        
# ловим ошибки        
    except AttributeError:
        data.split("\n\n\n")
        
    except Exception as e:
        print(e)
        
    except RuntimeError as f:
        raise ImportError(f)


for r, table, seat, handHiro, board, pref_call, pref_raises, pref_allin, \
    flop_check, flop_call, flop_bet, flop_raises, turn_call, turn_check, turn_bet, \
    turn_raises, river_call, river_check, river_bet, river_raises, showed_won in zip(result, Tables, 
                     seats, HandHiros, Boards, pref_calls, pref_raisess, pref_allins, flop_checks, 
                        flop_calls, flop_bets, flop_raisess, turn_calls, turn_checks, turn_bets, turn_raisess,
                                river_calls, river_checks, river_bets, river_raisess, showed_wons):
    
    #data_pd['TypeOfRoom'].append(r[0])
    data_pd['HandId'].append(r[1])
    data_pd['Tournament'].append(r[2])
    #data_pd['info'].append(r[3])
    data_pd['Level'].append(r[4])
    data_pd['Date'].append(r[5])
    data_pd['Hour'].append(r[6])
    data_pd['TimeZone'].append(r[7])
    data_pd['Players'].append(seat)
    data_pd['Table'].append(table) #table
    #data_pd['bain'].append(bain)
    data_pd['HandHiro'].append(handHiro[1])
    data_pd['Board'].append(board)
    data_pd['pref_call'].append(pref_call)
    data_pd['pref_raises'].append(pref_raises)
    data_pd['pref_allin'].append(pref_allin)
    data_pd['flop_call'].append(flop_call)
    data_pd['flop_check'].append(flop_check)
    data_pd['flop_bet'].append(flop_bet)
    data_pd['flop_raises'].append(flop_raises)
    data_pd['turn_call'].append(turn_call)
    data_pd['turn_check'].append(turn_check)
    data_pd['turn_bet'].append(turn_bet)
    data_pd['turn_raises'].append(turn_raises)
    data_pd['river_call'].append(river_call)
    data_pd['river_check'].append(river_check)
    data_pd['river_bet'].append(river_bet)
    data_pd['river_raises'].append(river_raises)
    data_pd['showed_won'].append(showed_won)

df_SNG_452883 = pd.DataFrame(data_pd)
df_SNG_452883.head()
```

```python
import math

def convert_size(size_bytes):  # размер файла
    if size_bytes == 0:
        return "0B"
    size_name = ("B", "KB", "MB", "GB", "TB", "PB", "EB", "ZB", "YB")
    i = int(math.floor(math.log(size_bytes, 1024)))
    p = math.pow(1024, i)
    s = round(size_bytes / p, 2)
    return "%s %s" % (s, size_name[i])


convert_size(df_54133.__sizeof__())
```
